# Часть 1. Преобразование в эмбеддинги

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
```

Оригинальный датасет:

```python
df_raw = pd.read_csv("h_raw_data.1500k.records.csv")
df_raw.info()
```

Столбики, которые будем склеивать в единый текст:

```python
features = ['country_title', 'city_title', 'about', 'activities', 'books', 'games', 'interests']
```

Чтобы их склеить без проблем, нужно запонить пропуски:

```python
for col in features:
    df_raw[col].fillna('', inplace=True)
df_raw.info()
```

Собственно само склеивание в тексты:

```python
texts = []
for i in range(len(df_raw)):
    res = ''
    for col in features:
        if df_raw.loc[i, col] == '':
            continue
        else:
            res += f"{df_raw.loc[i, col]}, "
    texts.append(res)
texts
```

Пример результата:

```python
texts[891509]
```

Если нужно добавить такой столбик, то делаем:

df_raw["text"] = texts

Преобразуем такие тексты в эмбеддинги:

```python
embedding_model = SentenceTransformer("paraphrase-multilingual-mpnet-base-v2") # Сюда можно вставлять любой другой sentence transformer
embeddings = embedding_model.encode(texts, show_progress_bar=True)
embeddings
```

Размер одного эмебеддинга:

```python
len(embeddings[0])
```

Если нужно добавить такой столбик, то делаем:

df_raw["embedding"] = embeddings

# Часть 2. Создание нового признака (номер кластера)

```python
from bertopic import BERTopic
from bertopic.representation import KeyBERTInspired, MaximalMarginalRelevance, PartOfSpeech
from hdbscan import HDBSCAN
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from umap import UMAP
```

```python
umap_model = UMAP(
    n_neighbors=5,
    n_components=3,
    min_dist=0.0,
    metric='cosine',
    random_state=42
)
```

```python
hdbscan_model = HDBSCAN(
    min_cluster_size=5,
    metric='chebyshev',
    cluster_selection_method='leaf',
    prediction_data=True
)
```

```python
stopwords_ru = stopwords.words("russian")
stopwords_en = stopwords.words("english")
stop_words = stopwords_ru + stopwords_en
stop_words
```

```python
vectorizer_model = CountVectorizer(
    min_df=2,
    stop_words=stop_words
)
```

```python
keybert_model = KeyBERTInspired()
mmr_model = MaximalMarginalRelevance(diversity=0.3)
```

```python
representation_model = {
    "KeyBert": keybert_model,
    "MMR": mmr_model
}
```

```python
topic_model = BERTopic(
    embedding_model=embedding_model,
    umap_model=umap_model,
    hdbscan_model=hdbscan_model,
    vectorizer_model=vectorizer_model,
    representation_model=representation_model,
    top_n_words=10,
    verbose=True
)
topics, probs = topic_model.fit_transform(texts, embeddings)
```

```python
topic_model.generate_topic_labels(
    nr_words=3,
    topic_prefix=False,
    word_length=10,
    separator=", "
)
```
